Collect 5 random PTIDs and their SUVr for Dhruv to compare.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/huw_whole_df.csv')
df = df[df['DX_bl'] == 0]
```

```python
suvr_cols = ['PTID']+[col for col in df.columns if col.startswith("SUVr")]

df = df[suvr_cols].reset_index()

df
```

```python
idx = []
for i in range(0,5):
    n = np.random.randint(0,118)
    idx.append(n)
```

```python
idx
```

```python
valid_idx = [i for i in idx if i in df.index]
df = df.loc[valid_idx]
```

```python
df.to_csv('data/test_results.csv')
```

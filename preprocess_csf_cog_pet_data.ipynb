```python
dx_map = {
    'AD': 0,
    'Dementia': 0,
    'MCI': 1,
    'EMCI': 1,
    'LMCI': 1,
    'CN': 2,
    'SMC': 3
}
df['DX'] = df['DX'].map(dx_map)
df['DX_bl'] = df['DX_bl'].map(dx_map)
```

```python
df['DX'].value_counts()
```

```python
df['DX_bl'].value_counts()
```

```python
gender_map = {
    'Male': 0,
    'Female': 1
}
df['PTGENDER'] = df['PTGENDER'].map(gender_map)
```

```python
df['PTRACCAT'].value_counts()
```

```python
race_map = {
    'White': 0,
    'Black': 1,
    'Asian': 2,
    'More than on e': 3,
    'Am Indian/Alaskan': 4
}
df['PTRACCAT'] = df['PTRACCAT'].map(race_map)
```

```python
df['PTRACCAT'].value_counts() # Drop for now since it is heavily imbalanced
```

```python
# Extract numerical values from 'ABETA_bl' and 'PTAU_bl' columns with rows in format <num or >num
df['ABETA_bl'] = df['ABETA_bl'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
df['PTAU_bl'] = df['PTAU_bl'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
# Columns to be dropped
# cols_to_drop = ['PTRACCAT', 'EXAMDATE_bl', 'EXAMDATE', 'PTID']
cols_to_drop = ['EXAMDATE_bl', 'EXAMDATE', 'PTID']
df.drop(columns=cols_to_drop, inplace=True)
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(f"Non-numerical columns: {df.select_dtypes(exclude=['int64', 'float64']).columns.tolist()}")
```

```python
# Convert all na rows in FHQMOMAD and FHQDADAD to 0 since the question is not asked so the answer is no
df.fillna({'FHQMOMAD': 0, 'FHQDADAD': 0, 'FHQMOM': 0, 'FHQDADAD': 0}, inplace=True)
```

```python
# Drop rows with any missing values
df.dropna(inplace=True)
```

```python
df.info(verbose=True, show_counts=True)
```

```python
# Save
df.to_csv(os.path.join(data_dir, 'CSF_COG_PET_data_cleaned.csv'), index=False)
```

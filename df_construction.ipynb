```python
import pandas as pd
import numpy as np
import os
```

```python
data_dir = 'data'

df_1 = pd.read_csv(os.path.join(data_dir, "CSF_COG_PET_data.csv"))
df_2 = pd.read_csv(os.path.join(data_dir, "CSF_COG_PET_data_cleaned.csv"))
```

```python
df_1.head()
```

```python
dx_map = {
    'AD': 0,
    'Dementia': 0,
    'MCI': 1,
    'EMCI': 1,
    'LMCI': 1,
    'CN': 2,
    'SMC': 3
}
df_1['DX'] = df_1['DX'].map(dx_map)
df_1['DX_bl'] = df_1['DX_bl'].map(dx_map)

gender_map = {
    'Male': 0,
    'Female': 1
}
df_1['PTGENDER'] = df_1['PTGENDER'].map(gender_map)

race_map = {
    'White': 0,
    'Black': 1,
    'Asian': 2,
    'More than one': 3, # error here was on e not one
    'Am Indian/Alaskan': 4
}
df_1['PTRACCAT'] = df_1['PTRACCAT'].map(race_map)

df_1['ABETA_bl'] = df_1['ABETA_bl'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
df_1['PTAU_bl'] = df_1['PTAU_bl'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)

df_1.fillna({'FHQMOMAD': 0, 'FHQDADAD': 0, 'FHQMOM': 0, 'FHQDADAD': 0}, inplace=True)

```
